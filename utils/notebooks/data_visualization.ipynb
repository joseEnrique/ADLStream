```python
import arff
from scipy.io import arff
import pandas as pd
import numpy as np
import math


_RESULT_DIR = '../../datasets_arff/timeseries/'

data, meta = arff.loadarff(_RESULT_DIR +'ArrowHead.arff')
attributes = [x for x in meta]
print(attributes)
df = pd.DataFrame(data)

df.head()

```

```python
df.shape
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
s = df.iloc[1].values[:-1]
plt.figure()
plt.plot(s, color='black')
plt.show()

```

```python
s1 = [abs(x) for x in  s]

plt.figure()
plt.plot(s1, color='black')
plt.show()


s1 = [max(s1[i], s1[i+1]) for i in range(0, len(s1), 2) if i<len(s1)-1]

plt.figure()
plt.plot(s1, color='black')
plt.show()
```

```python
s2=[abs(x+0.5) for x in  s]

plt.figure()
plt.plot(s2, color='black')
plt.show()

s2 = [max(s2[i], s2[i+1]) for i in range(0, len(s2), 2) if i<len(s2)-1]

plt.figure()
plt.plot(s2, color='black')
plt.show()
```

```python
s3=[abs(x-0.5) for x in  s]
plt.figure()
plt.plot(s3, color='black')
plt.show()

s3 = [max(s3[i], s3[i+1]) for i in range(0, len(s3), 2) if i<len(s3)-1]

plt.figure()
plt.plot(s3, color='black')
plt.show()
```

```python
s11=[2*x1+(-1)*x2+(-1/5)*x3 for x1,x2,x3 in  zip(s1,s2,s3)]
plt.figure()
plt.plot(s11, color='black')
plt.show()

s11 = [max(s11[i], s11[i+1]) for i in range(0, len(s11), 2) if i<len(s11)-1]

plt.figure()
plt.plot(s11, color='black')
plt.show()

s12=[2*math.log(x1)+(-1/2)*(1.1**x2)+(-0.1)*x3 for x1,x2,x3 in  zip(s1,s2,s3)]
plt.figure()
plt.plot(s12, color='black')
plt.show()

s12 = [max(s12[i], s12[i+1]) for i in range(0, len(s12), 2) if i<len(s12)-1]

plt.figure()
plt.plot(s12, color='black')
plt.show()
```

```python
s21=[-0.5*x1+(+1)*x2+(-1.5)*x3 for x1,x2,x3 in  zip(s1,s2,s3)]
plt.figure()
plt.plot(s21, color='black')
plt.show()

s21 = [max(s21[i], s21[i+1]) for i in range(0, len(s21), 2) if i<len(s21)-1]

plt.figure()
plt.plot(s21, color='black')
plt.show()

s22=[-1.5*x1+(+1)*math.e**x2+(-0.5)*x3 for x1,x2,x3 in  zip(s1,s2,s3)]
plt.figure()
plt.plot(s22, color='black')
plt.show()

s22 = [max(s22[i], s22[i+1]) for i in range(0, len(s22), 2) if i<len(s22)-1]

plt.figure()
plt.plot(s22, color='black')
plt.show()
```

```python
s31=[(-1.66)*math.log(x1)+(-1.33)*x2+(1)*math.e**x3 for x1,x2,x3 in  zip(s1,s2,s3)]
plt.figure()
plt.plot(s31, color='black')
plt.show()

s31 = [max(s31[i], s31[i+1]) for i in range(0, len(s31), 2) if i<len(s31)-1]

plt.figure()
plt.plot(s31, color='black')
plt.show()

s32=[(-1)*abs(x1)+(-1)*abs(x2)+(10)*abs(x3) for x1,x2,x3 in  zip(s1,s2,s3)]
plt.figure()
plt.plot(s32, color='black')
plt.show()

s32 = [max(s32[i], s32[i+1]) for i in range(0, len(s32), 2) if i<len(s32)-1]

plt.figure()
plt.plot(s32, color='black')
plt.show()
```

```python
s111=[ x11+x12+x21+x22+x31+x32 for x11,x12,x21,x22,x31,x32 in  zip(s11,s12,s21,s22,s31,s32)]
s111 = pd.Series(s111).rolling(7).sum()
plt.figure()
plt.plot(s111, color='black')
plt.show()

s111 = [max(s111[i], s111[i+1]) for i in range(0, len(s111), 2) if i<len(s111)-1]

plt.figure()
plt.plot(s111, color='black')
plt.show()

s111=[ x11*x12*x21*x22*x31*x32 for x11,x12,x21,x22,x31,x32 in  zip(s11,s12,s21,s22,s31,s32)]
s111 = pd.Series(s111).rolling(7).mean()
plt.figure()
plt.plot(s111, color='black')
plt.show()

s111 = [max(s111[i], s111[i+1]) for i in range(0, len(s111), 2) if i<len(s111)-1]

plt.figure()
plt.plot(s111, color='black')
plt.show()


s111=[ x11+math.e**x12+x21+(x22)+x31+math.log(abs(x32)) for x11,x12,x21,x22,x31,x32 in  zip(s11,s12,s21,s22,s31,s32)]
s111 = pd.Series(s111).rolling(7).mean()
plt.figure()
plt.plot(s111, color='black')
plt.show()

s111 = [max(s111[i], s111[i+1]) for i in range(0, len(s111), 2) if i<len(s111)-1]

plt.figure()
plt.plot(s111, color='black')
plt.show()

s111=[ math.log(abs(x11+x12+x21))+math.e**(x22)-x31-x32 for x11,x12,x21,x22,x31,x32 in  zip(s11,s12,s21,s22,s31,s32)]
s111 = pd.Series(s111).rolling(7).mean()
plt.figure()
plt.plot(s111, color='black')
plt.show()

s111 = [max(s111[i], s111[i+1]) for i in range(0, len(s111), 2) if i<len(s111)-1]

plt.figure()
plt.plot(s111, color='black')
plt.show()

s111=[ (abs(x11+x12+x21+x22+x31+x32))**(1/2) for x11,x12,x21,x22,x31,x32 in  zip(s11,s12,s21,s22,s31,s32)]
s111 = pd.Series(s111).rolling(7).mean()
plt.figure()
plt.plot(s111, color='black')
plt.show()

s111 = [max(s111[i], s111[i+1]) for i in range(0, len(s111), 2) if i<len(s111)-1]

plt.figure()
plt.plot(s111, color='black')
plt.show()

s111=[x11**1*x12**2*x21**4 + x22**2*x31**2*x32*2 for x11,x12,x21,x22,x31,x32 in  zip(s11,s12,s21,s22,s31,s32)]
s111 = pd.Series(s111).rolling(7).mean()
plt.figure()
plt.plot(s111, color='black')
plt.show()

s111 = [max(s111[i], s111[i+1]) for i in range(0, len(s111), 2) if i<len(s111)-1]

plt.figure()
plt.plot(s111, color='black')
plt.show()

s111=[ np.min([x11,x12,x21,x22,x31,x32]) for x11,x12,x21,x22,x31,x32 in  zip(s11,s12,s21,s22,s31,s32)]
s111 = pd.Series(s111).rolling(7).mean()
plt.figure()
plt.plot(s111, color='black')
plt.show()

s111 = [max(s111[i], s111[i+1]) for i in range(0, len(s111), 2) if i<len(s111)-1]

plt.figure()
plt.plot(s111, color='black')
plt.show()
```
